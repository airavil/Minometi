```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel("Ballistic data for M821 HE shell (4 Rings).xlsx")
```

```python
Range=data['Range']
Elev=data['Elev']
```

```python
x=Range
y=Elev
plt.scatter(x, y, marker='.', linewidths=1, color='red')
plt.legend()
plt.title('Elev(Range)')
plt.xlabel("Range, M")
plt.ylabel("Elev, Mil")
plt.grid(True)
plt.show()
```

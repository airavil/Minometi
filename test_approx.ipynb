```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel("Ballistic data for M821 HE shell (4 Rings).xlsx")
```

```python
Range=data['Range']
Elev=data['Elev']
```

```python
x=Range/1000
y=np.cos(Elev/1000)
coeffs = np.polyfit(x, y, 1)
poly_func = np.poly1d(coeffs)
x_fit = np.linspace(0, 2.9, 3000)
y_fit = poly_func(x_fit)
# plt.plot(x_fit, y_fit, color='blue', label=str(poly_func))
plt.plot(x, y, color='blue', label=str(poly_func))
plt.scatter(x, y, marker='.', linewidths=1, color='red')
plt.legend()
plt.title('Elev(Range)')
plt.xlabel("Range, KM")
plt.ylabel("Elev, Rad")
plt.grid(True)
plt.show()
```
